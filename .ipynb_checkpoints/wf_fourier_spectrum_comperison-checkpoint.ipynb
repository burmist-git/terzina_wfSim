```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

%matplotlib inline
```

```python
def fourier_ana(csv_file):
    tic = time.time()
    #
    df=pd.read_csv(csv_file,index_col='time_ns')
    #
    dt=(df.index[1]-df.index[0])
    sig_time_window_width = (df.index[-1]-df.index[0])
    charge=float(df.sum()*sig_time_window_width)
    #
    FFT = np.fft.fft(df['v_sig'].values)
    N=len(FFT)
    N_fft=int(len(FFT)/2)
    unit_frequency=1.0/(df.index[-1]-df.index[0])/1.0e-9/1.0e6 # in MHz
    #
    frequency=unit_frequency*np.arange(0,N_fft)
    ampl=np.abs(FFT[:N_fft])
    df_fourier_real=pd.DataFrame(data=FFT[:N_fft].real,
                               index=frequency,
                               columns=['real'])
    df_fourier_imag=pd.DataFrame(data=FFT[:N_fft].imag,
                               index=frequency,
                               columns=['imag'])
    df_fourier=pd.DataFrame(data=ampl,
                            index=frequency,
                            columns=['ampl'])
    df_fourier=pd.concat([df_fourier_real,df_fourier_imag,df_fourier],axis=1)
    #
    cos_sin_df_l=[]
    for k in range(0,N_fft):
        colname_cos = str('cos_' + str(k))
        colname_sin = str('sin_' + str(k))
        cos_df = pd.DataFrame(data=FFT[k].real*np.cos(2*np.pi*k*df.index.values/dt/N),
                              index=df.index.values,
                              columns=[colname_cos])
        sin_df = pd.DataFrame(data=-FFT[k].imag*np.sin(2*np.pi*k*df.index.values/dt/N),
                              index=df.index.values,
                              columns=[colname_sin])
        cos_sin_df_l.append(pd.concat([cos_df,sin_df],axis=1))
    #
    df_cos_sin=pd.concat(cos_sin_df_l,axis=1)
    #
    df_cos_sin_copy = df_cos_sin.copy()
    df_cos_sin_copy.reset_index(inplace=True,drop=True)
    df_cos_sin_copy=df_cos_sin_copy.transpose()
    #df_cos_sin_copy.to_csv(str(csv_file + '.cos_sin.csv'),sep=' ')
    #
    print('N = ',N)
    print('N_fft = ',N_fft)
    print('unit_frequency in MHz = ', unit_frequency)
    toc = time.time()
    print('Execution time:', toc-tic, 'seconds')
    #
    return df, df_fourier, df_cos_sin_copy
```

```python
def filter_signal(df, df_cos_sin, cut_off_frequency_min, cut_off_frequency_max):
    #
    dt_sig=(df.index[1]-df.index[0])
    sig_time_window_width = (df.index[-1]-df.index[0]) 
    charge=float(df.sum()*sig_time_window_width) 
    #
    cut_off_ind_min = int(cut_off_frequency_min/dt_sig)
    cut_off_ind_max = int(cut_off_frequency_max/dt_sig)
    #
    df_sig_filter = df_cos_sin[cut_off_ind_min:cut_off_ind_max].sum(axis=0)
    #
    sig_filter_time = df_sig_filter.index*dt_sig
    sig_filter_amplitude = df_sig_filter.values-np.mean(df_sig_filter.values[:100])
    sig_filter_amplitude = sig_filter_amplitude/float(sig_filter_amplitude.sum()*sig_time_window_width)*charge
    #
    return pd.DataFrame(data=sig_filter_amplitude,
                        index=sig_filter_time,
                        columns=['ampl_filter'])
```

```python
df_25um, df_fourier_25um, df_cos_sin_25um=fourier_ana(csv_file='./wfData/waveform_template_fast_25um_lee.csv')
df_25um_conv, df_fourier_25um_conv, df_cos_sin_25um_conv=fourier_ana(csv_file='./wfData/waveform_template_fast_25um_lee_conv.csv')
df_35um, df_fourier_35um, df_cos_sin_35um=fourier_ana(csv_file='./wfData/waveform_template_fast_35um_lee.csv')
df_35um_conv, df_fourier_35um_conv, df_cos_sin_35um_conv=fourier_ana(csv_file='./wfData/waveform_template_fast_35um_lee_conv.csv')
#
df_gaus_short, df_fourier_gaus_short, df_cos_sin_gaus_short=fourier_ana(csv_file='./wfData/waveform_template_gaus_short.csv')
df_gaus_long, df_fourier_gaus_long, df_cos_sin_gaus_long=fourier_ana(csv_file='./wfData/waveform_template_gaus_long.csv')
#
#df_25um_conv_20MHz, df_fourier_25um_conv_20MHz, df_cos_sin_25um_conv_20MHz=fourier_ana(csv_file='./wfData/waveform_template_fast_25um_lee_conv_cut_off_frequency_20MHz.csv')
```
